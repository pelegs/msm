```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.set_cmap('rainbow')
```

```python
from numpy import sqrt, exp, log, cosh, tanh, pi
from numpy import abs as npabs
```

```python
def U(x, m, s):
    return -log(exp(-(m-x)**2/(2*s**2)) + exp(-(m+x)**2/(2*s**2)))

def dU(x, m, s):
    return 1/s**2 * (x-m*tanh(x*m/s**2))

def d2U(x, m, s):
    return 1/s**2 - (m/(s**2*cosh(m*x/s**2)))**2

def rate(m, s):
    dE = -log(2) + m**2/2 + log(1+exp(-2*m**2/2))
    wa = d2U(m, m, s)
    wb = npabs(d2U(0, m, s))
    return sqrt(wa*wa)/(2*pi) * exp(-dE)
```

```python
xs = np.arange(-10, 10, 0.01)
plt.plot(xs, U(xs, 3, 1))
plt.plot(xs, dU(xs, 3, 1))
plt.plot(xs, d2U(xs, 3, 1))
```

```python
ms = np.arange(0, 7, 0.01)
plt.plot(ms, rate(ms, 1))
```
