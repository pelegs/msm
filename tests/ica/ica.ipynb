```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.set_cmap('rainbow')
```

```python
x = np.random.normal(size=(1000,2))
plt.plot(x[:,0], x[:,1], '.')
```

```python
A = np.array([[2,3],
              [2,1]])
s = np.dot(A, x.T)
plt.plot(s.T[:,0], s.T[:,1], '.')
```
