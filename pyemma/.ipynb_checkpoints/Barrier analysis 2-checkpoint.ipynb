```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pyemma
```

```python
data = np.load('../data/dx_3.00.npy')
data = data.T[0]
zeros = np.zeros(data.shape)
data = np.vstack((data, zeros)).T
```

```python
cluster = pyemma.coordinates.cluster_kmeans(data, k=2, max_iter=50)
```

```python
%%capture
lags = np.array([5000, 7500, 10000, 50000])
its = pyemma.msm.its(cluster.dtrajs, lags=lags, nits=3, errors='bayes')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 3))
pyemma.plots.plot_feature_histograms(data, feature_labels=['$x$', '$y$'], ax=axes[0])
axes[1].set_xlabel('$x$')
axes[1].set_ylabel('$y$')
axes[1].set_xlim(-7, 7)
axes[1].set_ylim(-7, 7)
axes[1].set_aspect('equal')
print(type(axes[2]))
pyemma.plots.plot_implied_timescales(its, ylog=False, ax=axes[2])
fig.tight_layout()
```
