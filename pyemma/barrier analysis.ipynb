```python
import sys
sys.path.append('../lib')
from simlib import *

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pyemma
from scipy.stats import linregress
```

```python
num_sims = 40
ms = np.linspace(0.5, 4.0, num_sims)
Amp = np.ones(shape=(num_sims, 2, 2))
Mu = np.array([np.array([[-m, 0], [m, 0]]) for m in ms])
Sig = np.ones(shape=(num_sims, 2, 2))
Us = [potential(A, M, S) for A, M, S in zip(Amp, Mu, Sig)]
```

```python
params = []
for m, U in zip(ms, Us):
    params.append({
                    'name': 'libtest1',
                    'num_steps': 1000000,
                    'num_dim': 2,
                    'beta': 1,
                    'Ddt': 0.01,
                    'x0': np.array([m, 0]),
                    'potential': U
                    })
```

```python
Xs = []
for i, (p, m) in enumerate(zip(params, ms)):
    print('Now simulating for m={} ({} of {})'.format(m, i+1, len(ms)))
    trajectory = simulate(p)
    Xs.append(trajectory)
```

```python
clusters, its = [], []
tmax = [100*np.exp((m)**2/2) for i, m in enumerate(ms)]
for i, trajectory in enumerate(Xs):
    print('Now calculating {} of {}'.format(i+1, len(Xs)))
    clusters.append(pyemma.coordinates.cluster_kmeans(trajectory, k=2, max_iter=100))
    lags = np.linspace(1, tmax[i], 20).astype(int)
    its.append(pyemma.msm.its(clusters[i].dtrajs, lags=lags, nits=1, errors='bayes'))
```

```python
def display(data, cluster, its, m):
    fig, axes = plt.subplots(1, 1, figsize=(12, 3))
    pyemma.plots.plot_implied_timescales(its, ylog=False, ax=axes)
    plt.title('m={:0.2f}'.format(m))
    fig.tight_layout()
```

```python
for xs, cluster, timescales, m in zip(Xs, clusters, its, ms):
    display(xs, cluster, timescales, m)
```

```python
ts = [np.max(t.timescales) for t in its]
dGs = np.array([m**2/2 for m in ms])
M, b, r, _, _ = linregress(dGs, np.log(ts))
print('y={:0.3f}x+{:0.3f}, R^2={:0.3f}'.format(M, b, r**2))
theory = dGs * M + b
```

```python
from matplotlib import rc
rc('text', usetex=True)
plt.figure(figsize=(20,10))
plt.plot(dGs, theory)
plt.plot(dGs, np.log(ts), 'o')
plt.xlabel(r'$\Delta G \left(=\frac{\mu^{2}}{2}\right)$')
plt.ylabel(r'$\log \tau$')
plt.title(r'Time scales vs. Barrier height')
plt.savefig('plot.png')
```
