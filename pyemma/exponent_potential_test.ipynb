```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
class gaussian:
    def __init__(self, m, s, A):
        self.m = m
        self.s = s
        self.A = A
        
    def get_val(self, x):
        return self.A / np.sqrt(2*np.pi*self.s**2) * np.exp(-(x-self.m)**2/(2*self.s**2))
```

```python
def get_potential(xs, gs):
    ys = np.zeros(len(xs))
    for g in gs:
        ys += np.array([g.get_val(x) for x in xs])
    return -np.log(ys)
```

```python
xs = np.arange(-10, 10, 0.01)
well_center = np.arange(1, 5, 0.5)
dE = []
for a in well_center:
    g1 = gaussian(-a, 1/a, 1)
    g2 = gaussian(+a, 1/a, 1)
    U = get_potential(xs, [g1, g2])
    plt.plot(xs, U)
    #mid_point = len(xs)//2
    #dE.append(U[mid_point] - np.min(U))
```
