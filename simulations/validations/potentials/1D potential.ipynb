```python
import numpy as np
from simlib import *
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.set_cmap('rainbow')
```

```python
gaussians = []
gaussians.append(gaussian(A = np.ones(1),
                          M = np.array([-3]),
                          S = np.ones(1)))
gaussians.append(gaussian(A = np.ones(1),
                          M = np.array([3]),
                          S = np.ones(1)))
```

```python
xs = np.arange(-10, 10, 0.1)
U = potential(gaussians)
ys = [U.get_value([x]) for x in xs]
Fs = [U.get_force([x]) for x in xs]
plt.plot(xs, ys)
plt.plot(xs, Fs)
```
