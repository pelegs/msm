```python
import numpy as np
from scipy.stats import linregress
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.set_cmap('rainbow')
```

```python
N = 242
ms = np.zeros(N)
ts = np.zeros(N)
errs = np.zeros(N)
with open('timescales.data', 'r') as f:
    for i, line in enumerate(f):
        ms[i], ts[i], errs[i] = [float(x) for x in line.rstrip().split(' ')]
```

```python
xs = ms**2/2 + np.log(2)
log_its = np.log(ts)
```

```python
m_min, m_max = 75, -1
slope, intercept, r, _, _, = linregress(xs[m_min:m_max], log_its[m_min:m_max])
print(slope, intercept, r**2)
ys = slope * xs + intercept
plt.errorbar(xs[m_min:m_max], log_its[m_min:m_max], yerr=np.sqrt(errs/ts)[m_min:m_max], fmt='.')
plt.plot(xs[m_min:m_max], ys[m_min:m_max])
```
