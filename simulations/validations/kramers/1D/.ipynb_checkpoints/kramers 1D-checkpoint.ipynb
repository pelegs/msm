```python
import numpy as np
from simlib import *
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.set_cmap('rainbow')
```

```python
import pyemma
```

```python
# Potential
m = 2.40
gaussians = []
gaussians.append(gaussian(A=np.ones(1),
                          M=np.array([-m]),
                          S=np.ones(1)))
gaussians.append(gaussian(A=np.ones(1),
                          M=np.array([m]),
                          S=np.ones(1)))
U = potential(gaussians)

# Parameters
parameters = {
    'name': 'equilibrium_test',
    'num_steps': 1000000,
    'num_dim': 1,
    'num_particles': 1,
    'KBT': 1,
    'Ddt': 0.01,
    'x0': 0.0,
    'potential': U
}

# Simulation
fname = 'm_{:0.2f}.npy'.format(m)
Xs = np.load(fname)

# PyEmma
fig, axes = plt.subplots(1, 1, figsize=(10, 5))
cluster = pyemma.coordinates.cluster_kmeans(Xs, k=2, max_iter=50)
its = pyemma.msm.its(cluster.dtrajs, lags=np.arange(1,2500,250), nits=1, errors='bayes')
axes.set_title('m={:0.2f}'.format(m))
pyemma.plots.plot_implied_timescales(its, ylog=False, ax=axes)
fig.tight_layout()
```
