```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.set_cmap('rainbow')
import glob
```

```python
Xs_all = np.empty(2)
for path in glob.glob('data/*'):
    data = np.loadtxt(path)
    Xs_all = np.vstack((Xs_all, data))
```

```python
xbins = np.arange(-7, 7, 0.1)
ybins = np.arange(-7, 7, 0.1)
```

```python
fig, ax = plt.subplots()
hist = ax.hist2d(Xs_all[:,0], Xs_all[:,1], bins=(xbins, ybins))
```
