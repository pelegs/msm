```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def gauss(x, m=0, s=1):
    return np.exp(-(x-m)**2/(2*s**2))
```

```python
def g(x):
    return gauss(x, -3.5, 2) + gauss(x, 3.5, 2) + gauss(x, 0.0, 1.0)
```

```python
xs = np.linspace(-7, 7, 1000)
```

```python
p = g(xs)
U = -np.log(p)
```

```python
#plt.plot(xs, p)
plt.plot(xs, U)
```
