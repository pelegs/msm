```python
from matplotlib import pyplot as plt 
import numpy as np  
%matplotlib inline
```

```python
import sys
sys.path.append('../lib')
from simlib import *
```

```python
ts, xs, bins, bin_edges = simulate(method='langevin', num_particles=10000, max_t=10)
```

```python
sim_MSD, sim_stdev = MSD(xs)
```

```python
plt.plot(ts[:-1], sim_MSD)
plt.plot(ts[:-1], sim_stdev)
```

```python
ms1, covs1 = [], []
ms2, covs2 = [], []
Ds = np.arange(0.1, 5.5, 0.5)
for D in Ds:
    ts1, xs1, _, _ = simulate(method='langevin', num_particles=10000, max_t=50, D=D)
    ts2, xs2, _, _ = simulate(method='smol', num_particles=10000, max_t=50, D=D)
    sim_MSD1, sim_stdev1 = MSD(xs1)
    sim_MSD2, sim_stdev2 = MSD(xs2)
    m1, cov1 = np.polyfit(ts1, sim_MSD1, 1, cov=True)
    m2, cov2 = np.polyfit(ts2, sim_MSD2, 1, cov=True)
    ms1.append(m1)
    ms2.append(m2)
    covs1.append(cov1)
    covs2.append(cov2)
```

```python
plt.plot(Ds, ms1, 'o')
plt.plot(Ds, ms2, '*')
```
