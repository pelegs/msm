```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def gauss(x, m, s):
    return np.exp(-(x-m)**2/(2*s**2))
```

```python
N = 7
xmin, xmax = [-15, 15]
m = np.random.uniform(xmin, xmax, size=N)
s = np.random.uniform(1, 3, size=N)


xs = np.linspace(xmin-3, xmax+3, 1000)
ps = [gauss(xs, mi, si) for mi, si in zip(m, s)]
p = np.sum(ps, axis=0)
U = -np.log(p)

plt.plot(xs, U)
```
